```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('/content/breast-cancer-wisconsin.data',header=None)
```

```python
df.head()
```

```python
col_names={0:'id',
1:'Clump Thickness',
2:'Uniformity of Cell Size',
3:'Uniformity of Cell Shape',
4:'Marginal Adhesion',
5:'Single Epithelial Cell Size',
6:'Bare Nuclei',
7:'Bland Chromatin',
8:'Normal Nucleoli',
9:'Mitoses',
10:'Class'}
```

```python
df.rename(columns=col_names,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
dt=df.copy() #copying the dataframe to retrieve
```

```python
dt.head()
```

```python
df['Bare Nuclei'].value_counts()
```

```python
df['Bare Nuclei']=df['Bare Nuclei'].replace('?',0)
```

```python
df['Bare Nuclei']=df['Bare Nuclei'].replace(0,df['Bare Nuclei'].astype(int).mean())
```

```python
df['Bare Nuclei'].value_counts()
```

```python
df['Bare Nuclei']=df['Bare Nuclei'].astype(int)
```

```python
df=df.drop('id',axis=1)
```

```python
x=df[df.columns[:-1]]
y=df['Class']
```

```python
x.head()
```

```python
y.head()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
```

```python
pd.DataFrame(predictions).value_counts()
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,predictions)
```

```python
mat=metrics.confusion_matrix(y_test,predictions)
```

```python
mat_df=pd.DataFrame(mat,index=['benign','malignant'],columns=['predict_Benign','predict_Malignant'])
mat_df
```

```python
sns.heatmap(mat_df,annot=True,cmap='viridis')
```

**stratifying the train test data for better fit**

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x,y,test_size=0.25,stratify=y)
```

```python
y1_train.value_counts()
```

```python
model1=GaussianNB()
```

```python
model1.fit(x1_train,y1_train)
```

```python
predict1=model1.predict(x1_test)
```

```python
pd.DataFrame(predict1).value_counts()
```

```python
metrics.accuracy_score(y1_test,predict1)
```

```python
con_mat=metrics.confusion_matrix(y1_test,predict1)
con_mat
```

```python
con_mat_df=pd.DataFrame(con_mat,index=['benign','malignant'],columns=['predict_Benign','predict_Malignant'])
con_mat_df
```

```python
sns.heatmap(con_mat_df,annot=True)
plt.show()
```

**Here the above model is able to predict the all Malignants as malignants, but few cases which are benign are also predicted as malignant.**

**In this case our model could able to predict the all the malignants which is prefered in medical industry, with accuracy morethan 98 percent**
